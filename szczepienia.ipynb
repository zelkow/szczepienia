
## Loading the data from cvs files


```python
# file downloaded to /home/jovyan/work/data/
file = "/home/jovyan/work/data/poziomszcz.csv"


# loading file to dataframe
df = pd.read_csv(file,encoding= 'utf8',sep=';', encoding_errors='ignore')

#let's check what we have in the file
df.info()
df
```

```python
# Let's group by region and sort by population
```

```python
#df.groupby('wojewodztwo_nazwa')

```

```python
df.groupby(['wojewodztwo_nazwa']).sum()

```

```python
wojewod = df.groupby('wojewodztwo_nazwa').sum().sort_values(by=['wojewodztwo_nazwa'])
wojewod['pc'] = (wojewod['w3_zaszczepieni_pelna_dawka'] / wojewod['liczba_ludnosci']) * 100
wojewod
```

```python
z_wojewod = wojewod['pc'].sort_values(ascending=True)
print(z_wojewod)
type(z_wojewod)
```

```python
plt.figure(figsize=(10, 10))
z_wojewod.plot(kind='barh')
plt.show
```

```python
powiat = df.groupby('powiat_nazwa').sum().sort_values(by=['liczba_ludnosci'])
powiat['pc'] = (powiat['w3_zaszczepieni_pelna_dawka'] / powiat['liczba_ludnosci']) * 100
```

```python
powiat.info()
```

```python
z_powiat = (powiat['w3_zaszczepieni_pelna_dawka'] / powiat['liczba_ludnosci'])  * 100
z_powiat = z_powiat.sort_values(ascending=True)

plt.figure(figsize=(10, 100))
z_powiat.plot(kind='barh')
plt.show
```
